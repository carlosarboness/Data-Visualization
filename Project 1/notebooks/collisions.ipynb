```python
import pandas as pd
import altair as alt
```

```python
collisions = pd.read_csv("../data/preprocessed-colisions.csv")
 collisions.head()
# len(collisions)
```

## Visualize the data

### Visualization 1

```python
df1 = pd.DataFrame({'CRASH DATE': pd.to_datetime(collisions["CRASH DATE"])})
df1['day'] = df1['CRASH DATE'].dt.day_name()
df1['year'] = df1['CRASH DATE'].dt.year
df1 = df1.groupby(['day', 'year']).mean().reset_index(names=['day', 'year'])
df1
```

```python
chart1 = alt.Chart(df1).mark_bar().encode(
    x='year:O',
    y='counts:Q',
    color='year:N',
    column=alt.Column('day:N', title='Day of the Week', 
                            sort=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

)

chart1
```

## Visualization 3

```python
df3 = pd.DataFrame({'CRASH TIME': pd.to_datetime(collisions["CRASH TIME"]), 
                    'CRASH DATE': pd.to_datetime(collisions["CRASH DATE"])})

df3 = pd.DataFrame({'hour': (df3['CRASH TIME'].apply(lambda x: str(x.hour))),
                    'year': df3['CRASH DATE'].dt.year})

df3['hour'] = df3['hour'].apply(lambda x: int(x))

df3 = df3.groupby(['hour', 'year']).size().reset_index(name='counts')
df3.head()
```

```python
alt.Chart(df3).mark_line().encode(
    x='hour:O',
    y='counts:Q',
    color='year:N'
)
```

## VISUALIZATION 5

```python
df5 = pd.DataFrame({'datetime': pd.to_datetime(collisions["CRASH DATE"])})
df5 = df5.groupby(['datetime']).size().reset_index(name='collisions')
#check the collisions type of data
df5.dtypes
df5.head()
```

```python
weather2018 = pd.read_csv("weather2018.csv")
weather2020 = pd.read_csv("weather2020.csv")
weather = pd.concat([weather2018, weather2020])
weather['datetime'] = pd.to_datetime(weather['datetime'])
weather.head()
```

```python
df5 = pd.merge(df5, weather, on='datetime')
df5.head()
```

```python
alt.Chart(df5).mark_point().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='year:Q'
).repeat(
    row = ['collisions'],
    column = ['temperature', 'humidity', 'wind_speed']
).properties(
    width=150,
    height=150
)
```
