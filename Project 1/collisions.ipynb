```python
import pandas as pd
import altair as alt
```

```python
collisions = pd.read_csv("preprocessed-colisions.csv")
collisions.head()
```

## Visualize the data

### Visualization 1

```python
df1 = pd.DataFrame({'CRASH DATE': pd.to_datetime(collisions["CRASH DATE"])})
df1['day'] = df1['CRASH DATE'].dt.day_name()
df1['year'] = df1['CRASH DATE'].dt.year
df1 = df1.groupby(['day', 'year']).size().reset_index(name='counts')
df1
```

```python
chart1 = alt.Chart(df1).mark_bar().encode(
    x='year:O',
    y='counts:Q',
    color='year:N',
    column=alt.Column('day:N', title='Day of the Week', 
                            sort=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

)

chart1
```

## Visualization 3

```python
df3 = pd.DataFrame({'CRASH TIME': pd.to_datetime(collisions["CRASH TIME"]), 
                    'CRASH DATE': pd.to_datetime(collisions["CRASH DATE"])})

df3 = pd.DataFrame({'hour': (df3['CRASH TIME'].apply(lambda x: str(x.hour))),
                    'year': df3['CRASH DATE'].dt.year})

df3['hour'] = df3['hour'].apply(lambda x: int(x))

df3 = df3.groupby(['hour', 'year']).size().reset_index(name='counts')
df3.head()
```

```python
alt.Chart(df3).mark_line().encode(
    x='hour:O',
    y='counts:Q',
    color='year:N'
)
```
